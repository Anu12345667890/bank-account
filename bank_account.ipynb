```python
balance = 0

def create():
    global balance
    print("The account is created")

def deposit():
    global balance
    amount = float(input("Enter the amount to be deposited: "))
    balance += amount
    print("The deposit is successful, and the balance in the account is %.2f" % balance)

def withdraw():
    global balance
    amount = float(input("Enter the amount to withdraw: "))
    if balance >= amount:
        balance -= amount
        print("The withdrawal is successful, and the balance is %.2f" % balance)
    else:
        print('Insufficient Balance')

def enquiry():
    global balance
    print("Balance in the account is %.2f" % balance)

while True:
    print("\nWhat do you want?")
    print("Create account: 1\nDeposit: 2\nWithdraw: 3\nEnquiry: 4\nExit: 5")
    choice = int(input("Your choice: "))

    if choice == 1:
        create()
    elif choice == 2:
        deposit()
    elif choice == 3:
        withdraw()
    elif choice == 4:
        enquiry()
    elif choice == 5:
        print("Exiting the program.")
        break
    else:
        print("Invalid choice")

    another_operation = input("Do you want anything else? (yes/no): ")
    if another_operation.lower() != 'yes':
        print("Exiting the program.")
        break
```
